```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can also try "gpt2-medium" for a larger model
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Function to generate content using GPT-2
def generate_niche_content(prompt, max_length=200):
    # Encode the prompt into tokens
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text from the model
    outputs = model.generate(
        inputs,
        max_length=max_length,  # Maximum length of the generated text
        num_return_sequences=1,  # Number of sequences to generate
        no_repeat_ngram_size=2,  # Avoid repeating the same n-grams
        do_sample=True,  # Sampling for randomness
        temperature=0.7,  # Sampling temperature to increase creativity
        top_k=50,  # Consider top K tokens by probability
        top_p=0.95  # Nucleus sampling (top-p sampling)
    )

    # Decode the generated tokens to text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Ask user to input their custom prompt
niche_prompt = input("Enter the product description or niche market prompt: ")

# Call the function to generate content based on user input
generated_content = generate_niche_content(niche_prompt)

# Display the generated content
print("\nAI-Generated Content:")
print(generated_content)
```
