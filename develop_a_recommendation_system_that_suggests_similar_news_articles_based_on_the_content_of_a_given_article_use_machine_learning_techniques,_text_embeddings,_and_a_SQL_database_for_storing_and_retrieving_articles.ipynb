```python
import numpy as np
import torch
from sentence_transformers import SentenceTransformer, util

# Step 1: Initialize the Embedding Model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight model for embedding generation.

# Step 2: Define Database Connection (actual database used)
DATABASE_URL = "postgresql+psycopg2://username:password@localhost/news_db"  # URL for placeholder purposes.

# Step 3: In-Memory Data Structure to Simulate Database
articles = []  # List to store articles with their embeddings.

# Step 4: Helper Function to Add Articles
def insert_article(title, content):
    embedding = model.encode(content).tolist()  # Generate embedding for the content
    articles.append({
        "id": len(articles) + 1,  # Auto-increment ID
        "title": title,
        "content": content,
        "embedding": np.array(embedding),  # Store embedding as a NumPy array
    })

# Step 5: Add Sample Articles
sample_articles = [
    {"title": "Renewable Energy Adoption in Europe Surges in 2024",
     "content": "Europe has seen a dramatic increase in renewable energy adoption this year, driven by government incentives and technological advancements."},
    {"title": "How Government Policies Are Driving Solar Energy Growth",
     "content": "Government initiatives worldwide have significantly boosted the adoption of solar energy. This article explores key policy changes in the last decade."},
    {"title": "The Role of Wind Energy in Achieving Carbon Neutrality",
     "content": "Wind energy plays a pivotal role in the global transition to renewable sources. This article examines its impact and future potential."},
    {"title": "Technological Breakthroughs in Renewable Energy for 2024",
     "content": "New technologies are revolutionizing the renewable energy sector. Learn about the latest advancements in solar panels and wind turbines."},
    {"title": "AI in Healthcare: Transforming Diagnosis and Treatment",
     "content": "Artificial intelligence is revolutionizing healthcare, with advancements in diagnostics, treatment planning, and patient care."},
    {"title": "The Future of Electric Vehicles: Challenges and Opportunities",
     "content": "The electric vehicle industry faces challenges like charging infrastructure but also offers opportunities for sustainable transportation."},
    {"title": "Space Exploration: The Quest to Colonize Mars",
     "content": "NASA and private companies are making strides in space exploration, focusing on technologies needed to colonize Mars."},
    {"title": "Advancements in Quantum Computing: A New Era of Technology",
     "content": "Quantum computing is set to revolutionize industries, from cryptography to materials science, by solving problems classical computers can't."},
    {"title": "The Rise of Remote Work: How Companies Are Adapting",
     "content": "Remote work is here to stay, with businesses adopting flexible work policies and investing in collaboration tools."},
    {"title": "Climate Change and Its Impact on Global Agriculture",
     "content": "Global warming is affecting agriculture, with changing weather patterns disrupting food production and threatening food security."},
]

for article in sample_articles:
    insert_article(article['title'], article['content'])

# Step 6: Recommendation Function
def recommend_similar_articles(input_article_content, top_k=5):
    input_embedding = model.encode(input_article_content)  # Encode the input article
    input_embedding = torch.tensor(input_embedding, dtype=torch.float32)  # Ensure float32 for compatibility

    # Compute similarities
    similarities = []
    for article in articles:
        article_embedding = torch.tensor(article["embedding"], dtype=torch.float32)  # Ensure float32
        similarity = util.cos_sim(input_embedding, article_embedding).item()  # Compute cosine similarity
        similarities.append((article["title"], similarity))

    # Sort and return top-k results
    similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[:top_k]
    return similarities

# Step 7: Get User Input and Test Recommendation
print("Enter an article content to find similar articles:")
user_input_article = input("> ")  # Prompt the user for article content

# Get recommendations based on user input
recommendations = recommend_similar_articles(user_input_article)

# Display recommendations
print("\nRecommended Articles:")
for title, score in recommendations:
    print(f"Title: {title} | Similarity Score: {score:.4f}")
```
