```python
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer

# Load model and tokenizer
model_name = "gpt2-medium"  # You can try "gpt-3.5-turbo" if you have OpenAI API access.
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Function to generate content based on platform length requirements
def generate_adapted_content(content, platform):
    platform_specs = {
        "blog": {"max_length": 400, "temperature": 0.7},
        "social_media": {"max_length": 100, "temperature": 0.9},
        "email": {"max_length": 150, "temperature": 0.8},
    }

    # Adjust parameters based on platform
    specs = platform_specs.get(platform, platform_specs["blog"])

    # Generate content
    generated_content = generator(
        content,
        max_length=specs["max_length"],
        temperature=specs["temperature"],
        top_k=50,
        top_p=0.95,
        do_sample=True,
        num_return_sequences=1
    )

    return generated_content[0]["generated_text"]

# User input for base content and platform
base_content = input("Enter the base content you want to adapt: ")
platform = input("Enter the target platform (blog, social_media, email): ").strip().lower()

# Generate and display adapted content
if platform in ["blog", "social_media", "email"]:
    adapted_content = generate_adapted_content(base_content, platform)
    print(f"\n--- {platform.capitalize()} Version ---\n{adapted_content}")
else:
    print("Invalid platform. Please enter 'blog', 'social_media', or 'email'.")
```
