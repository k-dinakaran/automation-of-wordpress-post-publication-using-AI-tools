```python
# Install required libraries
# pip install transformers
# pip install yake  # For keyword extraction

import re
from transformers import pipeline
from yake import KeywordExtractor

# Step 1: Initialize Sentiment Analysis Model
sentiment_analyzer = pipeline("sentiment-analysis")

# Step 2: Initialize Keyword Extractor
keyword_extractor = KeywordExtractor(lan="en", n=2, dedupLim=0.9, top=3, features=None)

# Step 3: Social Media Preprocessing Function
def preprocess_social_media_post(post):
    # Remove hashtags, mentions, URLs, and special characters
    post = re.sub(r'@\w+', '', post)  # Remove mentions
    post = re.sub(r'#\w+', '', post)  # Remove hashtags
    post = re.sub(r'http\S+|www\S+', '', post)  # Remove URLs
    post = re.sub(r'[^\w\s]', '', post)  # Remove special characters
    return post.strip()

# Step 4: Function for Recommendations
def generate_recommendation(sentiment):
    if sentiment == "NEGATIVE":
        return "Consider escalating the issue for resolution or public response."
    elif sentiment == "POSITIVE":
        return "Highlight this feedback in campaigns or share it on your social platforms."
    else:
        return "Monitor further for additional context or sentiment trends."

# Step 5: Interactive User Input for Social Media Posts
print("Welcome to the Social Media Sentiment Analysis Tool!")
print("Enter your social media post or type 'exit' to quit.")
print("-" * 50)

while True:
    user_input = input("Your Post: ")

    # Exit condition
    if user_input.lower() == 'exit':
        print("Thank you for using the Social Media Sentiment Analysis Tool. Goodbye!")
        break

    # Step 6: Preprocess the Input
    cleaned_post = preprocess_social_media_post(user_input)

    # Step 7: Analyze Sentiment
    analysis = sentiment_analyzer(cleaned_post)
    sentiment = analysis[0]['label']
    confidence = round(analysis[0]['score'], 4)

    # Step 8: Extract Key Phrases
    keywords = keyword_extractor.extract_keywords(cleaned_post)
    key_phrases = [kw[0] for kw in keywords]  # Extract just the phrases

    # Step 9: Generate Recommendation
    recommendation = generate_recommendation(sentiment)

    # Step 10: Display Results
    print("\nAnalysis Result:")
    print(f"Original Post: {user_input}")
    print(f"Cleaned Post: {cleaned_post}")
    print(f"Sentiment: {sentiment} (Confidence: {confidence})")
    print(f"Key Phrases: {', '.join(key_phrases)}")
    print(f"Recommendation: {recommendation}")
    print("-" * 50)
```
